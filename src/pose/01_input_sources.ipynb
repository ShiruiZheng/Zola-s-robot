```python
import cv2

# -------------------------------
# Function to get video input from:
# 1. Laptop webcam
# 2. Local video file
# 3. Phone camera 
# -------------------------------
def get_video_capture(source="webcam", video_path=None):
    """
    Select a video source for OpenCV.
    
    Parameters:
        source (str): "webcam" or "video"
        video_path (str): path to a local video file
        
    Returns:
        cv2.VideoCapture object
    """
    
    if source == "webcam":
        print("üé• Using Mac webcam or iPhone (Continuity Camera)")
        return cv2.VideoCapture(0)   # Automatically uses whichever is set as default in macOS

    elif source == "video":
        if not video_path:
            raise ValueError("You selected 'video' but did not provide video_path.")
        print(f"üéû Using local video file: {video_path}")
        return cv2.VideoCapture(video_path)

    else:
        raise ValueError("Invalid source. Choose 'webcam' or 'video'.")
```

Example: Using Mac or iPhone Camera

```python
cap = get_video_capture(source="webcam")

while True:
    ret, frame = cap.read()
    if not ret:
        print("‚ùå Cannot read frame.")
        break

    cv2.imshow("Camera Test", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

Example: Using a Local Video File

```python
cap = get_video_capture(source="video", video_path="dog.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        print("üìå End of video or cannot read file.")
        break

    cv2.imshow("Video File Test", frame)

    if cv2.waitKey(30) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
